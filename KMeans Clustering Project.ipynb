## KMeans Cluster Creation 

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis = 1))
```

```python
kmeans.cluster_centers_
```

## Evaluation

##### We do have the labels for this project, so we take advantage of this to evaluate our clusters.

```python
def new_column(private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(new_column) # Create a new column 'Cluster' that returns 1 if it's Private School, 0 otherwise
```

```python
df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(df['Cluster'],kmeans.labels_))
print('\n')
print(classification_report(df['Cluster'],kmeans.labels_))
```

The result doesn't seem to be so bad considering the algorithm is purely using the features to cluster the universities into 2 distinct groups.
